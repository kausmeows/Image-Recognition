```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

```python
frame = cv2.imread("doctor_strange.jpg") #cv2 read bgr instead of rgb i.e. red and blue are reversed in this case.
```

```python
plt.imshow(frame)
```

```python
rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces = classifier.detectMultiScale(frame) # gives x, y, width and height
```

```python
face = faces[0]
```

```python
x, y, w, h = face
```

```python
face = rgb[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```

```python
out = cv2.rectangle(rgb, (x,y), (x+w, y+h), (0, 123, 255), 10) 

# cv2.rectangle(image, start_point, end_point, color, thickness)
# the color here will be the color of rectangle that comes around face.
```

```python
plt.imshow(out)
```
