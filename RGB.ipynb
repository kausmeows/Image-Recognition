```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
pil = Image.open("rgb.jpg")
pil
```

```python
"""" images are actually pixels in RGB format i.e. every pixel has three layers R, G, B which imply red, green
 and blue respectively."""

arr = np.array(pil)
arr.shape

# in the output we can see that this image of our dear favourite avenger Mr. Strange has 1700 pixels in rows,
# 2560 in columns and 3 depth layers of R, G, B.
```

```python
"""" a[start:stop]  # items start through stop-1
a[start:]      # items start through the rest of the array
a[:stop]       # items from the beginning through stop-1
a[:]           # a copy of the whole array """

cut = arr[:5, :5] # this means first 5 rows and first 5 columns
```

```python
plt.imshow(cut) # initial 5 by 5 pixels will be shown
```

```python
# So what if we want to get the layer RED ???
red = arr[:, :, 0]
green = arr[:, :, 1]
blue = arr[:, :, 2]
```

```python
plt.imshow(red) # the highlighted portion in this is the red part.
```

```python
plt.imshow(green) #the highlighted portion shows green part
```

```python
plt.imshow(blue) #the highlighted portion shows blue part
```
